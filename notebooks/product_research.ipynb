Быстрый анализ дневных метрик и аномалий.

```python
import pandas as pd
from pathlib import Path

daily_path = Path('data/processed/daily_metrics.csv')
anomaly_path = Path('data/processed/daily_metrics_with_anomalies.csv')

daily = pd.read_csv(daily_path) if daily_path.exists() else pd.DataFrame()
anomalies = pd.read_csv(anomaly_path) if anomaly_path.exists() else pd.DataFrame()

print('Daily shape:', daily.shape)
print('Anomalies shape:', anomalies.shape)
```

```python
if not daily.empty:
    display(daily.tail(10))

if not anomalies.empty:
    display(anomalies[anomalies['is_anomaly'] == 1])
```
